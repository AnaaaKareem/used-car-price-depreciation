# Import Libraries

```python
import pandas as pd
import numpy as np
import os
```

# Load Datasets

```python
data_path = '../data/processed/'
uk_data_path = os.path.join(data_path, '100,000 UK Used Car Data set.csv')
cars_data_path = os.path.join(data_path, 'CARS DATASET (Audi, BMW, Ford, Hyundai, Skoda, VW).csv')
andrei_data_path = os.path.join(data_path, 'Used Cars Dataset Andrei Novikov.csv')

print("Loading datasets...")
df_uk = pd.read_csv(uk_data_path)
df_cars = pd.read_csv(cars_data_path)
df_andrei = pd.read_csv(andrei_data_path)

print(f"UK Dataset Shape: {df_uk.shape}")
print(f"CARS Dataset Shape: {df_cars.shape}")
print(f"Andrei Dataset Shape: {df_andrei.shape}")
```

# Merge Datasets

```python
print("Merging datasets together...")
final_df = pd.concat([df_uk, df_cars, df_andrei], ignore_index=True)

# Drop any exact duplicates
print(f"Shape before dropping duplicates: {final_df.shape}")
final_df = final_df.drop_duplicates()
print(f"Shape after dropping duplicates: {final_df.shape}")

# Ensure price exists
final_df = final_df.dropna(subset=['price'])

final_df.head()
```

# Export file

```python
output_file = os.path.join(data_path, 'combined_final_dataset.csv')
print(f"Exporting to {output_file}...")
final_df.to_csv(output_file, index=False)
print("Export complete!")
```
