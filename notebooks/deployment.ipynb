## Final Report

```python
final_report = {
    "title": "Used Car Price Depreciation — Predictive Model",
    "executive_summary": (
        "This project developed a machine learning model to predict used car prices based on "
        "key vehicle attributes (make, model, year, mileage, transmission, fuel type, mpg, engine size). "
        "Three datasets were combined (UK Used Cars, Craigslist Vehicles, Andrei Novikov dataset) "
        "to create a comprehensive training set. Eight regression models were evaluated, with "
        "Random Forest Regression achieving the best performance (R² = 0.9474, MAE = 1966.5425)."
    ),
    "key_findings": [
        "The specific vehicle 'model' is overwhelmingly the strongest predictor of price, accounting for nearly 69% of the model's predictive power. This indicates buyers heavily value specific models over the overarching brand ('make', ~3%).",
        "Mileage (~14.5%) and Age (~6.1%) are the primary drivers of mechanical and temporal depreciation. Together with the car model, these three features explain nearly 90% of a vehicle's value.",
        "Random Forest Regression outperformed linear models because used car depreciation is highly non-linear; the rate at which a car loses value changes drastically depending on the specific model and mileage milestones.",
        "Electric Vehicles (EVs) require specialized data handling due to fundamentally different mechanical baselines (e.g., 0.0L engine sizes), which can break standard pricing pipelines if not isolated."
    ],
    "model_performance": {
        "best_model": "Random Forest Regression",
        "R²": 0.9474,
        "MAE": 1966.5425,
        "RMSE": 2863.0914,
        "MAPE": 0.0877
    },
    "recommendations": [
        "Deploy the model to the purchasing team to calculate the 'Maximum Trade-In Bid' by subtracting the target 10% profit margin and estimated reconditioning costs from the model's predicted retail price.",
        "Since specific models drive value retention more than overarching brands, inventory managers should use the model to source specific high-value retaining vehicles rather than buying broadly based on brand reputation.",
        "Standardize currencies at the data ingestion layer in future pipeline iterations for more accurate global pricing insights.",
        "Begin capturing specific EV battery health data (State of Health) in future datasets to improve EV-specific depreciation curves."
    ],
    "limitations": [
        "The model assumes 'average' physical condition. A major limitation is the absence of 'Vehicle Condition' or 'Accident History' data, which were excluded because they were not universally available across all merged datasets.",
        "Regional price bias exists because the UK and US markets have different pricing structures and currencies that were not normalized in real-time.",
        "The model relies on historical static data and does not account for sudden economic shocks for example, inflation, changing interest rates, or government EV tax credits.",
        "Missing data in the Craigslist dataset required imputing MPG and Engine Size with medians, which slightly dilutes the accuracy for specific trims.",
        "Since it is historical static data each dataset must be setup manually in terms of data understanding and preparation which takes a huge effort."
    ]
}
```
